```python
import numpy as np
import matplotlib
font = {'size': 29}
matplotlib.rc('font', **font)
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
vals = np.ones((256, 4))
vals[:, 0] = np.linspace(1, 140/256, 256)
vals[:, 1] = np.linspace(1, 21/256, 256)
vals[:, 2] = np.linspace(1, 21/256, 256)
#newcmp = ListedColormap(vals)
cmap = ListedColormap(plt.get_cmap('viridis').colors[::-1])
```

```python
def confuse_mat(mat, title, name, clim=(0, 100)):
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111)
    
    im = ax.imshow(mat, interpolation='nearest', cmap=cmap)
    bar = ax.figure.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    im.set_clim(clim[0], clim[1])
    plt.axis('off')
    
    thresh = 0.5 * (clim[1] + clim[0])
    for i in range(mat.shape[0]):
        for j in range(mat.shape[1]):
            ax.text(j, i, format(mat[i, j], ".2f"),
                    ha="center", va="center",
                   color="white" if mat[i, j] > thresh else "black")

    fig.tight_layout()
    plt.savefig(name)
    plt.show()
```

```python
mat = np.array([[66.36, 46.28],
                [62.50, 49.09]])
title = 'Hammering (Ours)'
name = 'hammering_mat_ours'
confuse_mat(mat, title, name, [20, 70])

mat = np.array([[59.49, 21.54],
                [39.66, 27.27]])
title = 'Hammering (Template)'
name = 'hammering_mat_tmatch'
confuse_mat(mat, title, name, [70, 20])
```
