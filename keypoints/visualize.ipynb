```python
import os
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
def visualize_keypoints(point_cloud,
                        keypoints,
                        prefix,
                        name,
                        plot_lim=2,
                        point_size=4):
    [grasp_point, funct_point, funct_vect] = keypoints
    fig = plt.figure(figsize=(6, 6))
    xs = point_cloud[:, 0]
    ys = point_cloud[:, 1]
    zs = point_cloud[:, 2]

    ax = fig.add_subplot(111)
    ax.scatter(xs, ys, s=point_size, c='#2e2d29')
    ax.set_axis_off()
    ax.grid(False)
    plt.axis('equal')
    plt.savefig(os.path.join(prefix, '{}_point_cloud.png'.format(name)))
    plt.show()

    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111)
    ax.scatter(xs, ys, s=point_size, c='#2e2d29')
    ax.set_axis_off()
    ax.grid(False)
    
    ax.scatter(grasp_point[:, 0],
               grasp_point[:, 1],
               s=point_size * 30,
               c='#8c1515')
    
    ax.scatter(funct_point[:, 0],
               funct_point[:, 1],
               s=point_size * 30,
               c='#8c1515')
    
    ax.scatter(funct_point[:, 0] + funct_vect[:, 0],
               funct_point[:, 1] + funct_vect[:, 1],
               s=point_size * 30,
               c='#8c1515')
    plt.axis('equal')
    plt.savefig(os.path.join(prefix, '{}_keypoints.png'.format(name)))
    plt.show()
    plt.close()
    return
```

```python
data = 'utils/baseline/data_hammer.hdf5'
f = h5py.File(data, 'r')
point_cloud = f['pos_point_cloud']
keypoints = f['pos_keypoints']

for index in range(100):
    visualize_keypoints(point_cloud[index], 
                        np.split(keypoints[index], [1, 2], axis=0), 
                        'visualize', 
                        index)
```
