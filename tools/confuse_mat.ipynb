```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mat = np.array([[51.26, 35.00, 41.41],
                [40.38, 35.45, 44.30],
                [41.84, 51.32, 47.83]])
classes = ['Easy', 'Moderate', 'Hard']
title = 'Generalization across Objects'

plt.figure(figsize=(2, 2))
fig, ax = plt.subplots()
im = ax.imshow(mat, interpolation='nearest', cmap=plt.get_cmap('Greys'))
bar = ax.figure.colorbar(im, ax=ax)
im.set_clim(0, 99)


ax.set(xticks=np.arange(mat.shape[1]),
       yticks=np.arange(mat.shape[0]),
       xticklabels=classes, yticklabels=classes,
       title=title,
       ylabel='Training Objects',
       xlabel='Testing Objects')

thresh = mat.max() / 2.
for i in range(mat.shape[0]):
    for j in range(mat.shape[1]):
        ax.text(j, i, format(mat[i, j], ".2f"),
                ha="center", va="center",
                color="white" if mat[i, j] > thresh else "black")

fig.tight_layout()
plt.savefig('confusion_mat.pdf')
plt.show()
```
