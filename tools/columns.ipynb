```python
import numpy as np
import matplotlib
font = {'size': 27}

matplotlib.rc('font', **font)
import matplotlib.pyplot as plt
```

```python
def plot_columns(results,
                 title='Title',
                 save='save.pdf',
                 ylim=(20, 80),
                 methods=['NPM', 'KPwoI', 'Ours'],
                 tool_diff=['Hard Tools', 'Easy Tools'],
                 bar_width=0.35, 
                 bar_color=['#d2c295', '#8c1515'],
                 bar_alpha=0.80):
    
    plt.style.use('seaborn-dark')
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111)
    
    ind = np.arange(len(methods))
    ax.bar(ind - bar_width/2, 
           results[:, 0], 
           bar_width, 
           label=tool_diff[0], 
           color=bar_color[0], 
           alpha=bar_alpha,
           edgecolor='k')
    
    ax.bar(ind + bar_width/2,           
           results[:, 1], 
           bar_width, 
           label=tool_diff[1], 
           color=bar_color[1], 
           alpha=bar_alpha,
           edgecolor='k')
    
    ax.set_ylabel('Task Success Rate (%)')
    ax.set_title(title, fontsize=32)
    ax.set_xticks(ind)
    ax.set_xticklabels(methods)
    ax.set_ylim(ylim)
    plt.legend(loc='upper left')
    fig.tight_layout()
    plt.savefig(save)
    plt.show()
```

```python
results = np.array([[39.93, 52.60],
                    [55.79, 51.33],
                    [61.98, 71.85]])

plot_columns(results, 'Easy Hammering', 'easy_hammering.png', (20, 90))
```
